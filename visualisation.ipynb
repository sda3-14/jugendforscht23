```python
import numpy as np
import math

f = open("Steam" , 'r')
read = f.readlines()
f.close()
loss = [0,0.5,1,1.5,2,2.5,3,3.5,4]
bandwidth = [250,500,750,1000]
ping = []

print(len(read)/4)
for i in range(int(len(read)/4)):
    ins = []
    for j in range(4):
        h = read[i*4+j].split(" ")[2]
        h = h[:-2]
        h = round(float(h),2)
        ins.append(h)
    ping.append(ins)
    print(ins)

xv , yv = np.meshgrid(bandwidth,loss)
```

## Plot

```python
from matplotlib import pyplot as plt
import matplotlib

bandwidth = np.array(bandwidth, dtype=float)
loss = np.array(loss, dtype=float)
ping = np.array(ping, dtype=float)

Z=ping
X, Y = np.meshgrid(bandwidth,loss)
color_dimension = X
minn, maxx = color_dimension.min(), color_dimension.max()
norm = matplotlib.colors.Normalize(minn, maxx)
m = plt.cm.ScalarMappable(norm=norm, cmap='jet')
m.set_array([])
C = np.linspace(5, -5, Z.size).reshape(Z.shape)
scamap = plt.cm.ScalarMappable(cmap='inferno')
fcolors = scamap.to_rgba(C)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, facecolors=fcolors, cmap='inferno')

ax.set_ylabel('Loss in %')
ax.set_xlabel('Bandwidth in Kb/s')
ax.set_zlabel('Latency in ms')
```

```python
import numpy as np
import math

f = open("KCP" , 'r')
read = f.readlines()
f.close()
loss = [0,0.5,1,1.5,2,2.5,3,3.5,4]
bandwidth = [250,500,750,1000]
ping = []

print(len(read)/4)
for i in range(int(len(read)/4)):
    ins = []
    for j in range(4):
        h = read[i*4+j].split(" ")[2]
        h = h[:-2]
        ins.append(h)
    ping.append(ins)
    print(ins)

xv , yv = np.meshgrid(bandwidth,loss)
```

## Plot

```python
from matplotlib import pyplot as plt


fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour(bandwidth,loss, ping, 50,lw=3,colors="k", linestyles="solid", alpha=0.5)
ax.set_ylabel('Loss')
ax.set_xlabel('Bandwidth')
ax.set_zlabel('Latency')
```
